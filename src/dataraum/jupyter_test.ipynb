````python
# Semantic annotations - THE KEY INPUT to business_meaning entropy
# These fields directly feed into the detector:
# - business_description, business_name, entity_type, confidence, business_concept
print("=== SEMANTIC ANNOTATIONS (kontobuchungen) ===")
display(
    sql("""
    SELECT
        c.column_name,
        sa.semantic_role,
        sa.entity_type,
        sa.business_name,
        sa.business_description,
        sa.business_concept,
        sa.confidence,
        sa.annotation_source
    FROM semantic_annotations sa
    JOIN columns c ON sa.column_id = c.column_id
    JOIN tables t ON c.table_id = t.table_id
    WHERE t.table_name = 'kontobuchungen'
""")
)
````

## Step 5: Entropy Objects (from Entropy Phase)

The `BusinessMeaningDetector` produces these scores from semantic annotations.

````python
# Entropy objects for semantic.business_meaning dimension
print("=== ENTROPY OBJECTS (semantic.business_meaning) ===")
display(
    sql("""
    SELECT
        eo.target,
        eo.score,
        eo.confidence,
        eo.evidence,
        eo.resolution_options
    FROM entropy_objects eo
    WHERE eo.dimension = 'business_meaning'
    AND eo.layer = 'semantic'
    ORDER BY eo.score DESC
""")
)
````

````python
# Parse the evidence JSON to see the score breakdown

df = sql("""
    SELECT eo.target, eo.score, eo.evidence
    FROM entropy_objects eo
    WHERE eo.dimension = 'business_meaning' AND eo.layer = 'semantic'
    ORDER BY eo.score DESC
""")

print("=== SCORE BREAKDOWN (from evidence JSON) ===")
for _, row in df.iterrows():
    evidence = json.loads(row["evidence"]) if row["evidence"] else []
    if evidence and len(evidence) > 0:
        ev = evidence[0]
        raw = ev.get("raw_metrics", {})
        comp = ev.get("score_components", {})
        print(f"\n{row['target']} (score: {row['score']:.3f})")
        print(f"  has_description: {raw.get('has_description')}")
        print(f"  has_business_name: {raw.get('has_business_name')}")
        print(f"  has_entity_type: {raw.get('has_entity_type')}")
        print(f"  semantic_confidence: {raw.get('semantic_confidence')}")
        print(f"  has_business_concept: {raw.get('has_business_concept')}")
        print(
            f"  → base_score: {comp.get('base_score')}, conf_factor: {comp.get('confidence_factor')}, ontology_bonus: {comp.get('ontology_bonus')}"
        )
````

## Step 6: Entropy Interpretations (from Interpretation Phase - LLM)

The LLM generates human-readable Actions from the entropy scores.

````python
# Entropy interpretations - LLM-generated actions
print("=== ENTROPY INTERPRETATIONS (kontobuchungen) ===")
display(
    sql("""
    SELECT
        ei.table_name,
        ei.column_name,
        ei.composite_score,
        ei.readiness,
        ei.explanation,
        ei.resolution_actions_json
    FROM entropy_interpretations ei
    WHERE ei.table_name = 'kontobuchungen'
""")
)
````

````python
# Parse resolution actions for the affected columns
affected_columns = [
    "ID des Rechnungslegungsbereichs",
    "Kontonummer des Kontos",
    "Belegdatum der Buchung",
    "Belegfeld 2",
]

print("=== RESOLUTION ACTIONS (from LLM interpretations) ===")
for col in affected_columns:
    result = sql(f"""
        SELECT resolution_actions_json
        FROM entropy_interpretations
        WHERE column_name = '{col}'
    """)
    if not result.empty:
        actions = (
            json.loads(result.iloc[0]["resolution_actions_json"])
            if result.iloc[0]["resolution_actions_json"]
            else []
        )
        print(f"\n{col}:")
        for action in actions:
            print(f"  [{action.get('priority', '?').upper()}] {action.get('action', '?')}")
            print(f"    {action.get('description', '')[:100]}...")
            print(
                f"    effort: {action.get('effort', '?')}, impact: {action.get('expected_impact', '')[:60]}..."
            )
````

# Complete Column Lineage View

A single joined query showing all pipeline stages per column.

````python
# Re-open connection
conn = sqlite3.connect(sqlite_path)

# Complete column lineage query - joins all pipeline stages
lineage_query = """
WITH base AS (
    -- Base: columns from kontobuchungen
    SELECT
        c.column_id,
        c.column_name,
        t.table_name
    FROM columns c
    JOIN tables t ON c.table_id = t.table_id
    WHERE t.table_name = 'kontobuchungen' and t.layer='typed'
),

-- Stage 1: Type Decisions
typing AS (
    SELECT
        td.column_id,
        td.decided_type AS resolved_type,
        td.decision_source AS type_source,
        td.decision_reason AS type_reason
    FROM type_decisions td
),

-- Stage 1b: Type Candidates (for parse_success_rate)
type_candidates_agg AS (
    SELECT
        td.column_id,
        tc.parse_success_rate
    FROM type_decisions td
    JOIN type_candidates tc ON td.column_id = tc.column_id
        AND td.decided_type = tc.data_type
),

-- Stage 2: Statistical Profiles
statistics AS (
    SELECT
        sp.column_id,
        sp.total_count,
        sp.null_count,
        sp.distinct_count,
        sp.null_ratio,
        sp.cardinality_ratio
    FROM statistical_profiles sp
),

-- Stage 2b: Statistical Quality Metrics (for outliers)
quality_metrics AS (
    SELECT
        sqm.column_id,
        sqm.iqr_outlier_ratio,
        sqm.benford_compliant,
        sqm.iqr_outlier_ratio,
        sqm.isolation_forest_anomaly_ratio,
        sqm.quality_data
    FROM statistical_quality_metrics sqm
),

-- Stage 3: Semantic Annotations
semantic AS (
    SELECT
        sa.column_id,
        sa.semantic_role,
        sa.entity_type,
        sa.business_name,
        sa.business_description,
        sa.business_concept,
        sa.confidence AS semantic_confidence,
        sa.annotation_source
    FROM semantic_annotations sa
),

-- Stage 4: Entropy Objects (pivoted by dimension)
entropy_business_meaning AS (
    SELECT
        REPLACE(REPLACE(eo.target, 'column:kontobuchungen.', ''), 'column:', '') AS column_name,
        eo.score AS entropy_business_meaning, eo.confidence AS business_meaning_confidence, eo.evidence AS business_meaning_evidence, eo.resolution_options AS business_meaning_resolution_options
    FROM entropy_objects eo
    WHERE eo.layer = 'semantic' AND eo.dimension = 'business_meaning'
),

entropy_units AS (
    SELECT
        REPLACE(REPLACE(eo.target, 'column:kontobuchungen.', ''), 'column:', '') AS column_name,
        eo.score AS entropy_units, eo.confidence AS units_confidence, eo.evidence AS units_evidence, eo.resolution_options AS units_resolution_options
    FROM entropy_objects eo
    WHERE eo.layer = 'semantic' AND eo.dimension = 'units'
),

entropy_temporal AS (
    SELECT
        REPLACE(REPLACE(eo.target, 'column:kontobuchungen.', ''), 'column:', '') AS column_name,
        eo.score AS entropy_temporal, eo.confidence AS temporal_confidence, eo.evidence AS temporal_evidence, eo.resolution_options AS temporal_resolution_options
    FROM entropy_objects eo
    WHERE eo.layer = 'semantic' AND eo.dimension = 'temporal'
),

entropy_dimensional AS (
    SELECT
        REPLACE(REPLACE(eo.target, 'column:kontobuchungen.', ''), 'column:', '') AS column_name,
        eo.score AS entropy_dimensional, eo.confidence AS dimensional_confidence, eo.evidence AS dimensional_evidence, eo.resolution_options AS dimensional_resolution_options
    FROM entropy_objects eo
    WHERE eo.layer = 'semantic' AND eo.dimension = 'dimensional'
),

entropy_derived AS (
    SELECT
        REPLACE(REPLACE(eo.target, 'column:kontobuchungen.', ''), 'column:', '') AS column_name,
        eo.score AS entropy_derived_values, eo.confidence AS derived_confidence, eo.evidence AS derived_evidence, eo.resolution_options AS derived_resolution_options
    FROM entropy_objects eo
    WHERE eo.layer = 'computational' AND eo.dimension = 'derived_values'
),

entropy_types AS (
    SELECT
        REPLACE(REPLACE(eo.target, 'column:kontobuchungen.', ''), 'column:', '') AS column_name,
        eo.score AS entropy_types, eo.confidence AS types_confidence, eo.evidence AS types_evidence, eo.resolution_options AS types_resolution_options
    FROM entropy_objects eo
    WHERE eo.layer = 'structural' AND eo.dimension = 'types'
),

entropy_outliers AS (
    SELECT
        REPLACE(REPLACE(eo.target, 'column:kontobuchungen.', ''), 'column:', '') AS column_name,
        eo.score AS entropy_outliers, eo.confidence AS outliers_confidence, eo.evidence AS outliers_evidence, eo.resolution_options AS outliers_resolution_options
    FROM entropy_objects eo
    WHERE eo.layer = 'value' AND eo.dimension = 'outliers'
),

-- Stage 5: Entropy Interpretations
interpretations AS (
    SELECT
        ei.column_name,
        ei.composite_score,
        ei.readiness,
        ei.explanation, ei.assumptions_json, ei.resolution_actions_json
    FROM entropy_interpretations ei
    WHERE ei.table_name = 'kontobuchungen'
    AND ei.column_name IS NOT NULL
)

-- Final joined result
SELECT
    b.column_name,

    NULL AS "Typing Phase",
    t.resolved_type,
    tc.parse_success_rate,

    NULL AS "Statistics Phase",
    s.null_ratio,
    s.cardinality_ratio,
    s.distinct_count,

    NULL AS "Quality Metrics (outlier input)",
    qm.iqr_outlier_ratio,
    qm.benford_compliant,
    qm.iqr_outlier_ratio,
    qm.isolation_forest_anomaly_ratio,
    qm.quality_data,

    NULL AS "Semantic Phase",
    sem.semantic_role,
    sem.entity_type,
    sem.business_name,
    sem.business_concept,
    SUBSTR(sem.business_description, 1, 50) AS business_desc_preview,
    sem.semantic_confidence,

    NULL AS "Entropy Scores",
    ROUND(ebm.entropy_business_meaning, 3) AS ent_business_meaning,
    ebm.business_meaning_confidence,
    ebm.business_meaning_evidence,
    ebm.business_meaning_resolution_options,
    ROUND(eu.entropy_units, 3) AS ent_units,
    eu.units_confidence,
    eu.units_evidence,
    eu.units_resolution_options,
    ROUND(et.entropy_temporal, 3) AS ent_temporal,
    et.temporal_confidence,
    et.temporal_evidence,
    et.temporal_resolution_options,
    ROUND(ed.entropy_dimensional, 3) AS ent_dimensional,
    ed.dimensional_confidence,
    ed.dimensional_evidence,
    ed.dimensional_resolution_options,
    ROUND(edr.entropy_derived_values, 3) AS ent_derived,
    edr.derived_confidence,
    edr.derived_evidence,
    edr.derived_resolution_options,
    ROUND(ety.entropy_types, 3) AS ent_types,
    ety.types_confidence,
    ety.types_evidence,
    ety.types_resolution_options,
    ROUND(eout.entropy_outliers, 3) AS ent_outliers,
    eout.outliers_confidence,
    eout.outliers_evidence,
    eout.outliers_resolution_options,

    NULL AS "Interpretation",
    ROUND(i.composite_score, 3) AS composite_score,
    i.explanation,
    i.assumptions_json,
    i.resolution_actions_json,
    i.readiness

FROM base b
LEFT JOIN typing t ON b.column_id = t.column_id
LEFT JOIN type_candidates_agg tc ON b.column_id = tc.column_id
LEFT JOIN statistics s ON b.column_id = s.column_id
LEFT JOIN quality_metrics qm ON b.column_id = qm.column_id
LEFT JOIN semantic sem ON b.column_id = sem.column_id
LEFT JOIN entropy_business_meaning ebm ON b.column_name = ebm.column_name
LEFT JOIN entropy_units eu ON b.column_name = eu.column_name
LEFT JOIN entropy_temporal et ON b.column_name = et.column_name
LEFT JOIN entropy_dimensional ed ON b.column_name = ed.column_name
LEFT JOIN entropy_derived edr ON b.column_name = edr.column_name
LEFT JOIN entropy_types ety ON b.column_name = ety.column_name
LEFT JOIN entropy_outliers eout ON b.column_name = eout.column_name
LEFT JOIN interpretations i ON b.column_name = i.column_name
"""

df_lineage = pd.read_sql_query(lineage_query, conn)
print(f"=== COMPLETE COLUMN LINEAGE ({len(df_lineage)} columns) ===")
display(df_lineage)
````

````python
# Styled view - highlight high entropy columns
def highlight_entropy(val):
    """Color code entropy values: green < 0.2, yellow 0.2-0.5, red > 0.5"""
    if pd.isna(val):
        return ""
    if val < 0.2:
        return "background-color: #90EE90"  # light green
    elif val < 0.5:
        return "background-color: #FFFF99"  # light yellow
    else:
        return "background-color: #FF9999"  # light red


# Apply styling to entropy columns
entropy_cols = [
    "ent_business_meaning",
    "ent_units",
    "ent_temporal",
    "ent_dimensional",
    "ent_derived",
    "ent_types",
    "composite_score",
]

styled_df = df_lineage.style.apply(
    lambda s: s.map(highlight_entropy),
    subset=[c for c in entropy_cols if c in df_lineage.columns],
)
display(styled_df)
````
